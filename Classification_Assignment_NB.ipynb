```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dependent=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,CategoricalNB,ComplementNB,GaussianNB
skf1 = StratifiedKFold(n_splits=10)
skf2 = StratifiedKFold(n_splits=10)
skf3 = StratifiedKFold(n_splits=10)
skf4 = StratifiedKFold(n_splits=10)
skf5 = StratifiedKFold(n_splits=10)
params1 = {}
params2 = {}
params3 = {}
params4 = {}
params5 = {}
nb1 = MultinomialNB()
nb2 = BernoulliNB()
nb3 = CategoricalNB()
nb4 = ComplementNB()
nb5 = GaussianNB()
gs1 = GridSearchCV(nb1, cv=skf1, param_grid=params1, return_train_score=True,scoring='f1_weighted')
gs2 = GridSearchCV(nb2, cv=skf2, param_grid=params2, return_train_score=True,scoring='f1_weighted')

gs4 = GridSearchCV(nb4, cv=skf4, param_grid=params4, return_train_score=True,scoring='f1_weighted')
gs5 = GridSearchCV(nb5, cv=skf5, param_grid=params5, return_train_score=True,scoring='f1_weighted')
gs1.fit(X_train,Y_train)
gs2.fit(X_train,Y_train)

gs4.fit(X_train,Y_train)
gs5.fit(X_train,Y_train)
```

```python
re=gs1.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
re
```

```python
grid_prediction_bernoilli=gs2.predict(X_test)

grid_prediction_complement=gs4.predict(X_test)
grid_prediction_gaussian=gs5.predict(X_test)
grid_prediction_multinomial=gs1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_ber=confusion_matrix(Y_test,grid_prediction_bernoilli)

cm_com=confusion_matrix(Y_test,grid_prediction_complement)
cm_gau=confusion_matrix(Y_test,grid_prediction_gaussian)
cm_mul=confusion_matrix(Y_test,grid_prediction_multinomial)
```

```python
print(cm_ber,"\n\n",cm_com,"\n\n",cm_gau,"\n\n",cm_mul)
```

```python
from sklearn.metrics import classification_report
clf_ber=classification_report(Y_test,grid_prediction_bernoilli)
clf_com=classification_report(Y_test,grid_prediction_complement)
clf_gau=classification_report(Y_test,grid_prediction_gaussian)
clf_mul=classification_report(Y_test,grid_prediction_multinomial)
```
