```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dependent=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
from sklearn.preprocessing import StandardScaler
scx=StandardScaler()
X_train=scx.fit_transform(X_train)
X_test=scx.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
param_grid={'weights':['uniform','distance'],'algorithm':['auto','ball_tree','kd_tree','brute'],'metric':['minkowski','cityblock','cosine','euclidean','haversine','l1','l2','manhattan','nan_euclidean']}
grid=GridSearchCV(KNeighborsClassifier(n_neighbors=7),param_grid,verbose=3,cv=5,n_jobs=1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
cl=grid.cv_results_
```

```python
print(cl)
```

```python
grid_prediction=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,grid_prediction)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf=classification_report(Y_test,grid_prediction)
```

```python
print(clf)
```

```python
from sklearn.metrics import f1_score
f1_weight=f1_score(Y_test,grid_prediction,average='weighted')
print("The parameters for best f1 score {}".format(grid.best_params_),f1_weight)
```

```python
from sklearn.metrics import roc_auc_score
roc=roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
